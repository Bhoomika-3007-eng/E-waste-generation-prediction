```
datasets = [ewaste, gni, internet, imports, gdp_growth, urban, population]
for df in datasets:
    df["Year"] = df["Year"].astype(str)
    df["Country"] = df["Country"].str.strip()
print("Year and Country columns standardized across all datasets.")
```

```
master = ewaste.copy()
master = master.merge(gni, on=["Country", "Year"], how="inner")
master = master.merge(internet, on=["Country", "Year"], how="inner")
master = master.merge(imports, on=["Country", "Year"], how="inner")
master = master.merge(gdp_growth, on=["Country", "Year"], how="inner")
master = master.merge(urban, on=["Country", "Year"], how="inner")
master = master.merge(population, on=["Country", "Year"], how="inner")
print ("merge completed successfully.")
print("final shape:",master.shape)
master.head()
```
